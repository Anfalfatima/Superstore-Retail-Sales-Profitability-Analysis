```python
# imputation apply only on columns
sns.scatterplot(x='Sales', y='Profit', data=df, hue='Category')
plt.title('total sales vs profit')
plt.xlabel('total sales')
plt.ylabel('profit')
plt.show()
```

```python
category_profit = df.gruopby('Category')['Profit'].sum()
```

```python
category_profit = df.groupby('Category')['Sales'].sum()
category_profit
```

```python
# Grouping apply on categorical data
category_profit = category_profit.sort_values(ascending=False)
category_profit
```

```python
category_profit = df.groupby('Category')['Profit'].mean().reset_index()
category_profit
```

```python
avg_profit = df.groupby('Category')['Profit'].mean().reset_index()
avg_profit
```

```python
sns.barplot(x='Category', y='Profit' , data= df, palette='magma')
plt.title('Average profit by Category')
plt.show()
```

```python
# corr means how one column is related to another columns(always on numeric values)
corr = df[['Sales','Quantity','Discount','Profit']].corr()
corr
```

```python
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation heatmap')
plt.show()
```

```python
sns.heatmap(corr, cmap='coolwarm')
plt.title('Correlation heatmap')
plt.show()
```
